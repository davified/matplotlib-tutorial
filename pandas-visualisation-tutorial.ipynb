# Data Visualisation with Pandas

```python
#  df.plot() as a useful exploratory tool for quick throwaway plots.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
pd.options.display.max_columns = 40

plt.style.use('seaborn')
```

```python
# Load bank marketing data
df = pd.read_csv('./data/bank-additional-full.csv', sep=';')
```

```python
df['outcome'] = (df['y'] == 'yes').astype(int)
```

```python
df.head()
```

## Visualize our data

```python
### plotting continuous variables as a Series (i.e. numbers) - with .plot()
```

```python
df[['cons.conf.idx']].plot()
```

```python
# Multiple plots
df[['cons.conf.idx', 'cons.price.idx']].plot()
```

```python
# histogram
df[['age']].hist()
```

```python
df['job'].value_counts().plot.bar()
```

```python
# Scatter plots
df.plot.scatter(x='age', y='outcome')
```

```python
# Scatter plots
df.plot.scatter(x='age', y='outcome', s=df['duration'], alpha=0.1)
```

```python
df['job'].value_counts().plot.area()
```

```python
# Multiple plots

df[['duration', 'pdays']].plot.hist(subplots=True, figsize=(12,8))
```
